# CSV

```python
# Reading different data sources with the help of pandas
```

```python
from io import StringIO,BytesIO
```

```python
data = ('col1,col2,col3\n'
          'x,y,1\n'
          'a,b,2\n'
           'c,d,3')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
# Reading from csv column
df = pd.read_csv(StringIO(data),usecols=['col1','col2'])
```

```python
print(df)
```

```python
df=pd.read_csv(StringIO(data), usecols=lambda x: x.upper() in ['COL1', 'COL3'])
```

```python
print(df)
```

```python
df.to_csv('Test2.csv')
```

```python
## Specifying columns data types

data = ('a,b,c,d\n'
            '1,2,3,4\n'
            '5,6,7,8\n'
            '9,10,11,12')
```

```python
print(data)
```

```python
df=pd.read_csv(StringIO(data),dtype=object)
```

```python
df
```

```python
df['a'][1]
```

```python
df=pd.read_csv(StringIO(data),dtype={'b':int,'c':np.float,'a':'Int64'})
```

```python
df
```

```python
df['a'][1]
```

```python
## check the datatype
df.dtypes
```

```python
## Index columns and training delimiters
```

```python
data = ('index,a,b,c\n'
           '4,apple,bat,5.7\n'
            '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data),index_col=0)
```

```python
pd.read_csv(StringIO(data),index_col=3)
```

```python
data = ('a,b,c\n'
           '4,apple,bat,\n'
            '8,orange,cow,')
```

```python
pd.read_csv(StringIO(data))
```

```python
pd.read_csv(StringIO(data),index_col=False)
```

```python
## Combining usecols and index_col
data = ( 'a,b,c, \n'
     '4,apple,bat,\n'
        '8,orange,cow')
```

```python
pd.read_csv(StringIO(data), usecols=['b', 'c'],index_col=False)
```

```python
## Quoting and Escape Characters¶. Very useful in NLP

data = 'a,b\n"hello, \\"Bob\\", nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
```
