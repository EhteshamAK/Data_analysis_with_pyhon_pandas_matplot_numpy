## Pivot table allows you to transform or reshape data - by converting rows to columns or columns to rows or vice versa

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r'C:\Users\Pc World Computers\Desktop\Pandas - Complete\Temperature.xlsx')
```

```python
df.head(10)
```

```python
df.pivot(index = "City", columns="Date")
```

```python
df.pivot(index = "City", columns="Date",values="Temperature")
```

```python
df.isnull().sum()
```

```python
df.pivot_table(index = "City", columns="Date")
```
