## Tutorial No 4 - Bar Chart

simple bar chart showing revenues of us major tech companies

```python
company = ["Google","Amazon","Microsoft","Meta","Ibm","Oracle","Saleforce"]
revenue = [100,240,150,90,60,41,15] # Amount is in US Billions
```

```python
x_position = np.arange(len(company))
```

```python
plt.xlabel("US Companies")
plt.ylabel("Revenue in Billions")
plt.title("Revenue of US Major Tech Companies")

plt.bar(x_position,revenue, label="revenue")
plt.legend()
plt.grid()
```

```python
plt.xlabel("US Companies")
plt.ylabel("Revenue in Billions")
plt.xticks(x_position,company)
plt.title("Revenue of US Major Tech Companies")

plt.bar(x_position,revenue, label="revenue")
plt.legend()
plt.grid()
```

### Multiple Bar Showing Revenue and Profit of US Major Tech Companies

```python
# Add new column profit
profit = [20,12,10,14,2,9,7]
```

```python
plt.bar(x_position-0.2,revenue, width=0.4, label="Revenue")
plt.bar(x_position+0.2,profit, width=0.4, label="Profit")
plt.xlabel("US Companies")
plt.ylabel("Revenue in Billions")
plt.xticks(x_position,company)
plt.title("Revenue of US Major Tech Companies")
plt.legend()
plt.grid()
```

Horizontal bar chart using barh function

```python
plt.barh(x_position,revenue, label="Revenue")

plt.yticks(x_position,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

```python
plt.barh(x_position-0.2,revenue,height=0.4,label="Revenue")
plt.barh(x_position+0.2,profit,height=0.4,label="Profit")
plt.yticks(x_position, company)
plt.ylabel("US Tech Comapnies")
plt.xlabel("Revenue(billions)")
plt.legend()
```

## Tutorial No 5 - Histograms

```python
blood_sugar = [100,80,90,120,200,180,150,89,92,97,110]
plt.hist(blood_sugar, rwidth=0.8, bins=3)
```

Histogram showing normal, pre-diabetic and diabetic patients distribution
 1. 80-100: Normal
 2. 100-125: Pre-diabetic
 3. 80-100: Diabetic

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,150,200], rwidth=0.95, color='g')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step')
```

Horizontal Orientation

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')
```

# Tutorial No 6 - Pie Chart

```python
expense_values = [16000,6000,3000,1000,5000,1500,11000]
expense_labels = ['Home_Rent','Bills','Internet','MobileLoad','Committee','BikeLoan','PersonalLoan']
plt.pie(expense_values,labels=expense_labels,shadow=True)
plt.axis("equal")
plt.show()
```

Show Percentage for every Pie and increase the size of pie using radius 

```python
plt.pie(expense_values,labels = expense_labels, shadow=True, autopct='%0.0f%%', radius=2)
plt.show()
```

```python
plt.pie(expense_values,labels = expense_labels, shadow=True, autopct='%0.0f%%', radius=2, 
        explode=[0.1,0,0.2,0.1,0.2,0.1,0.1])
plt.show()
```

Saving Chart

```python
plt.pie(expense_values,labels = expense_labels, shadow=True, autopct='%0.0f%%', radius=2, 
        explode=[0.1,0,0.2,0.1,0.2,0.1,0.1])
plt.show()
plt.savefig('Ehtisham_expense.jpg',bbox_inches='tight', pad_inches=5,)
```
