### Histograms

```python
plt.figure(figsize=(12,5),dpi=350)
bins_y = [1,3,5]
bins = [0,50000,100000,150000,200000,250000,300000,350000,500000]
plt.hist(bahl.Salary,bins=bins, color='seagreen')
plt.xticks(bins)
plt.yticks(bins_y)
plt.title("No of staff with Salary Range")
plt.xlabel("Salary in PKR")
plt.ylabel('No of Staff')
plt.show()
```

###  Pie Chart

```python
range1 = bahl.loc[bahl['Salary']>=100000].count()[0]
range1
range2 = bahl.loc[bahl['Salary']<=100000].count()[0]
range2
```

```python
labels = ["Salaries over 1 Lac", "Salaries less than 1 lac"]
colors = ['#34e5eb','#cc34eb']
plt.pie([range1,range2], colors = colors, labels = labels, autopct='%0.0f %%')
plt.title("Staff with Salaries Ratio")
plt.show()
```

```python
bahl.columns
```

```python
bahl.salary_pkr = [int(x.strip('_pkr'))if type(x)==str else x  for x in bahl['salary_pkr']]
```

```python
bahl.salary_pkr
```

```python
bahl.salary_pkr.dtype
```

```python
Under_50k = bahl.loc[bahl['salary_pkr']<50000].count()[0]
Over_50k = bahl[(bahl['salary_pkr']>=50000)&(bahl['salary_pkr'] <100000)].count()[0]
b100k_200k = bahl[(bahl['salary_pkr']>=100000)&(bahl['salary_pkr']<200000)].count()[0]
b200K_300k = bahl[(bahl['salary_pkr']>=200000)&(bahl['salary_pkr']<300000)].count()[0]
b300k_above = bahl.loc[bahl['salary_pkr']<=300000].count()[0]
```

```python
salaries = [Under_50k,Over_50k,b100k_200k,b200K_300k,b300k_above]
labels = ['Under_50k','Over_50k','b100k_200k','b200K_300k','b300k_above']
plt.style.use('ggplot')
plt.pie(salaries,labels=labels, autopct='%0.0f %%',explode=[0,0,0,0,0.1])
plt.show()
```
